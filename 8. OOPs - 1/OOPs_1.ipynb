### Constructor With Multiple Parameters

```python
class Student:
    def __init__(self, name, rno, marks):
        self.name = name
        self.rno = rno
        self.marks = marks
    def display(self):
        print('Name: {}, RollNo: {}, Marks: {}'.\
              format(self.name, self.rno, self.marks))

    
s1 = Student('abc', 1, 90)

s1.display()

s2 = Student('xyz', 2, 80)

s2.display()

'''INSTANCE variable should be used using 'self', 
whether inside constructor or method
'''
```

## Variables 

### Instance Variable

```python
class Student:
    def __init__(self, name, rno, marks):
        self.name = name
        self.rno = rno
        self.marks = marks
    
    def display(self):
        print('Name: {}, RollNo: {}, Marks: {},'.\
              format(self.name, self.rno, self.marks))
      
s1 = Student('abc', 1, 90)

print(s1.__dict__)

print(s1.name)

s1.display()

s2 = Student('xyz', 2, 80)

s2.display()
```

### Static Variable

```python
class Student:
    college = 'DIT'
    def __init__(self, name, rno, marks):
        self.name = name
        self.rno = rno
        self.marks = marks
    
    def display(self):
        print('Name: {}, RollNo: {}, Marks: {},'.\
              format(self.name, self.rno, self.marks), end=' ')
        print('College:', self.college)
      
s1 = Student('abc', 1, 90)

s1.display()

s2 = Student('xyz', 2, 80)

s2.display()
```

```python
# # Lets try to change the STATIC variable using s1

s1.college = 'UIT'

s1.display()

s2.display()
```

### Local Variable

```python
class Test:
    def m1(self):
        x = 10
        print(x)
    def m2(self):
        y = 20
        print(y)
        
t = Test()
t.m1()
t.m2()
```

## Setters & Getters

```python
class Student:
    # Note: __ (double underscore) as a prefix to variable denotes private variable in Python
    __college = 'DIT'
    def __init__(self, name, rno, marks):
        self.__name = name
        self.__rno = rno
        self.__marks = marks
    
    def display(self):
        print('Name: {}, RollNo: {}, Marks: {},'.format(self.__name, self.__rno, self.__marks), end=' ')
        print('College:', self.__college)
        
    def setCollege(self, cname):
        self.__college = cname
        
    def getName(self):
        return self.name
      
s1 = Student('abc', 1, 90)

s1.setCollege('UIT')

s1.display()

s2 = Student('xyz', 2, 80)

s2.display()
```
